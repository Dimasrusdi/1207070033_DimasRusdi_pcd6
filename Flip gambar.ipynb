```python
import numpy as np # data dalam bentuk array
import imageio #membaca, menulis, dan memproses berbagai jenis format file gambar
import matplotlib.pyplot as plt #grafik
```

```python
img = imageio.imread("aadim.jpg") # Membaca gambar
```

```python
img_height = img.shape[0] #height gambar
img_width = img.shape[1] #width gambar 
img_channel = img.shape[2] # Mengambil jumlah channel (warna) gambar 
img_type = img.dtype # Mengambil tipe data gambar menggunakan metode "dtype" dari NumPy
```

```python
img_flip_horizontal = np.zeros(img.shape, img_type) # untuk menyimpan gambar yang sudah diflip secara horizontal
img_flip_vertical = np.zeros(img.shape, img_type) # untuk menyimpan gambar yang sudah diflip secara vertikal
```

```python
for y in range(0, img_height): # Looping untuk setiap piksel pada dimensi vertikal
    for x in range(0, img_width):  # Looping untuk setiap piksel pada dimensi horizontal
        for c in range(0, img_channel): # Looping untuk setiap channel 
            img_flip_horizontal[y][x][c] = img[y][img_width-1-x][c] # Melakukan flipping horizontal pada gambar, dan piksel 
```

```python
plt.imshow(img_flip_horizontal) # gambar yang telah diflip secara horizontal
plt.title("Flip Horizontal") # judul pada gambar 
plt.show() # Menampilkan gambar ke layar
plt.imshow(img_flip_vertical) # Menampilkan gambar yang telah diflip secara vertikal
plt.title("Flip Vertical") # Memberi judul pada gambar yang ditampilkan
plt.show() # Menampilkan gambar ke layar
```
